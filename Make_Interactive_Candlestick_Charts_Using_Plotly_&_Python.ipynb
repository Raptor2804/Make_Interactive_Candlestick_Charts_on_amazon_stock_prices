```
#Description: This program shows how to create an interactive candlestick chart in python using Plotly
```

```
import plotly.graph_objects as go
import pandas as pd
```

```
#Load stock data
from google.colab import files
uploaded = files.upload()
```

```
#Store data in a data frame
df = pd.read_csv('AMZNtrain.csv')
#Set the index
df = df.set_index(pd.DatetimeIndex(df['Date'].values))
#Show data
df
```

```
#Create an interactive candlestick chart
figure = go.Figure(
    data=[
          go.Candlestick(
              x = df.index,
              low = df['Low'],
              high = df['High'],
              close = df['Close'],
              open=df['Open'],
              increasing_line_color='orange',
              decreasing_line_color='black',
          )
          ]
)
# figure.update_layout(xaxis_rangeslider_visible=False)
figure.update_layout(
    title='Amazon Price',
    yaxis_title='Amazon Sock Price USD ($)',
    xaxis_title='Date'
)

figure.show()
```
